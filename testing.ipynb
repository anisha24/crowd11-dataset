```python
import matplotlib.pyplot as plt

# Data
class_names = [
    "Gas Free", "Gas Jammed", "Laminar Flow", "Turbulent Flow", 
    "Crossing Flows", "Merging Flow", "Diverging Flow", 
    "Static Calm", "Static Agitated", "Interacting Crowd", "No Crowd"
]
labels = list(range(len(class_names)))
clips_obtained = [377, 480, 1037, 862, 517, 227, 189, 686, 351, 153, 340]

# Create bar chart
plt.figure(figsize=(10, 6))
plt.bar(labels, clips_obtained, color='skyblue')

# Adding labels
plt.xlabel('Class Name', fontsize=14)
plt.ylabel('# Clips Obtained', fontsize=14)
plt.title('Number of Clips Obtained for Each Class', fontsize=16)
plt.xticks(labels, class_names, rotation=45, ha='right', fontsize=10)

# Show plot
plt.tight_layout()
plt.show()
```
